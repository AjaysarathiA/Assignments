```python
# Create a function that tells elegibility of marriage for male and female according to their age limit like 21 for male and 18 for female
```

```python
def eligible(age,gender):
    if gender.lower()=="male" and age>=21:
     print("Eligible for marriage")
    elif gender.lower()=="female" and age>=18:
     print("Eligible for marriage")
    else:
        print("not eligible")
```

```python
age=int(input("Enter your age: "))
gender=input("Enter your gender: ")
eligible(age,gender)
```

```python
#print area and perimeter of triangle using class and functions
```

```python
class Triangle:
    def area(height,breadth):
        return (height*breadth)/2
    def perimeter(height1,height2,breadth):
        return (height1+height2+breadth)
        
```

```python
print("area: ",Triangle.area(32,34))
print("perimeter: ",Triangle.perimeter(2,4,4))
```
